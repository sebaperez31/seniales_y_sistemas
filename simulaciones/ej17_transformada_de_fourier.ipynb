Ejercicio 17 - Transformada de Fourier - Item c)

```python
import numpy as np
import matplotlib.pyplot as plt

# vamos a graficar la entrada en parte real e imaginaria, notar que es periodica con periodo N = 4
D = 30 # vamos a usar una duracion de entrada de 30 muestras
n = np.array(range(D))
x = np.exp(1j * (np.pi/2) * n)
plt.stem(n, np.real(x), linefmt='b-', label='Parte real de x[n]', markerfmt='bo', basefmt='gray')
plt.stem(n, np.imag(x), linefmt='r-', label='Parte Imaginaria de x[n]', markerfmt='ro', basefmt='gray')
plt.legend() 
plt.show()
```

```python
# ahora vamos a construir a la salida utilizando a ecuacion en diferencias
y = np.zeros_like(x)
# el primer valor la calculo a mano
y[0] = x[0]
for i in range(1,D) : 
    y[i] = (0.5) * y[i-1] + x[i]

plt.stem(n, np.real(y), linefmt='b-', label='Parte real de y[n]', markerfmt='bo', basefmt='gray')
plt.legend() 
plt.show() 

plt.stem(n, np.imag(y), linefmt='r-', label='Parte Imaginaria de y[n]', markerfmt='ro', basefmt='gray')
plt.legend() 
plt.show()        
```

```python
# se observa que n = 6 aproximadamente la respuesta transitoria finalizo y el sistema entra en regimen permanente
# ahora vamos a comparar con la salida permamente obtenida en forma analitica
y_permanente = (4/5 - 1j*(2/5))*x

plt.stem(n, np.real(y), linefmt='b-', label='Parte real de y[n]', markerfmt='bo', basefmt='gray')
plt.stem(n, np.real(y_permanente), linefmt='r-', label='Parte real de y_permanente[n]', markerfmt='ro', basefmt='gray')
plt.legend() 
plt.show()

plt.stem(n, np.imag(y), linefmt='b-', label='Parte imaginaria de y[n]', markerfmt='bo', basefmt='gray')
plt.stem(n, np.imag(y_permanente), linefmt='r-', label='Parte imaginaria de y_permanente[n]', markerfmt='ro', basefmt='gray')
plt.legend() 
plt.show()
```

```python
# se observa que a partir de n = 5 aproximadamente y[n] es igual a la respuesta permanente obtenida en forma analitica (los graficos se superponen) 
```
